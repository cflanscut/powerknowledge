```python
import sys
sys.path.append('/home/chaofan/powerknowledge/data/')
from read_PLAID_data import read_processed_data,get_feature_name
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import tree
import pydotplus
from IPython.display import Image, display

x, y = read_processed_data('type',
                           selected_label=[
                               'Laptop', 'Compact Fluorescent Lamp',
                               'Incandescent Light Bulb'
                           ],
                           direaction=1,
                           offset=30,
                           each_lenth=1)
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.5,
                                                    random_state=0)
test_tree = tree.DecisionTreeClassifier(max_depth=2,min_samples_leaf=50)
test_tree.fit(x_train, y_train)
print(test_tree.score(x_test, y_test))

feature_name = get_feature_name()

dot_data = tree.export_graphviz(test_tree,
                                feature_names=feature_name,
                                class_names=['Compact Fluorescent Lamp','Incandescent Light Bulb','Laptop'],
                                filled=True,
                                rounded=True,
                                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
display(Image(graph.create_png()))
```
